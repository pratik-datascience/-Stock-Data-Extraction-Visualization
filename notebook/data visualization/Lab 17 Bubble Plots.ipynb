### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
df_bubble =df[['Age', 'SOPartFreq', 'JobSat']].dropna() 

age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
freq_map = {
    'Multiple times per day': 5,
    'Daily or almost daily': 4,
    'A few times per week': 3,
    'A few times per month or weekly': 2,
    'Less than once per month or monthly': 1,
    'I have never participated in Q&A on Stack Overflow': 0
}

df_bubble['Agenum'] = df_bubble['Age'].map(age_map) 
df_bubble['Freqnum'] =df_bubble['SOPartFreq'].map(freq_map) 


plt.figure(figsize=(10, 6))

# Scatter plot (bubble plot)
plt.scatter(
    df_bubble['Agenum'],
    df_bubble['Freqnum'],
    s=df_bubble['JobSat'] * 100,  # Scale bubble size
    c=df_bubble['JobSat'],        # Use JobSatNum for color
    cmap='viridis',                  # Color map
    alpha=0.6,
    edgecolors='w',
    linewidth=0.5
)

# Add labels and title
plt.xlabel('Age (Numeric)')
plt.ylabel('Participation Frequency')
plt.title('Age vs. Participation Frequency (Bubble Size = Job Satisfaction)')

# Add a colorbar to show job satisfaction levels
cbar = plt.colorbar()
cbar.set_label('Job Satisfaction (1=Low, 5=High)')

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Drop missing values
df_plot = df[['ConvertedCompYearly', 'JobSat', 'Age']].dropna().copy()

# Step 2: Map Age
age_map = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}
df_plot['AgeNum'] = df_plot['Age'].map(age_map)

# Step 3: Keep JobSat as-is (already numeric)
df_plot['JobSatNum'] = df_plot['JobSat']  # 🔥 this is the fix

# # Step 4: Filter by 85th percentile
# q95 = df_plot['ConvertedCompYearly'].quantile(0.95)
# df_plot = df_plot[df_plot['ConvertedCompYearly'] <= q95]

# Step 5: Drop any remaining missing AgeNum
df_plot = df_plot.dropna(subset=['AgeNum', 'JobSatNum', 'ConvertedCompYearly'])

print("Plotting rows:", df_plot.shape[0])

# Step 6: Plot
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    x=df_plot['ConvertedCompYearly'],
    y=df_plot['JobSatNum'],
    s=df_plot['AgeNum'] * 10,
    c=df_plot['AgeNum'],
    cmap='viridis',
    alpha=0.6,
    edgecolors='k'
)

plt.title('Compensation vs. Job Satisfaction\n(Bubble Size = Age)', fontsize=14)
plt.xlabel('Yearly Compensation (USD)', fontsize=12)
plt.ylabel('Job Satisfaction Score (1 = Low, 10 = High)', fontsize=12)
cbar = plt.colorbar(scatter)
cbar.set_label('Respondent Age (Numeric)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1–4: Preprocessing
df_lang = df[['LanguageHaveWorkedWith', 'Age']].dropna().copy()

age_map = {
    'Under 18 years old': 'Under 18',
    '18-24 years old': '18-24',
    '25-34 years old': '25-34',
    '35-44 years old': '35-44',
    '45-54 years old': '45-54',
    '55-64 years old': '55-64',
    '65 years or older': '65+'
}
df_lang['AgeGroup'] = df_lang['Age'].map(age_map)
df_lang['LanguageList'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_lang = df_lang.explode('LanguageList')

# Count total languages to get top 10
top_langs = df_lang['LanguageList'].value_counts().head(10).index

# Group by language and age
lang_counts = df_lang.groupby(['LanguageList', 'AgeGroup']).size().reset_index(name='Frequency')

# Filter to top 10 languages only
lang_counts_top10 = lang_counts[lang_counts['LanguageList'].isin(top_langs)]

# Plot
plt.figure(figsize=(14, 8))
scatter = plt.scatter(
    x=lang_counts_top10['LanguageList'],
    y=lang_counts_top10['AgeGroup'],
    s=lang_counts_top10['Frequency'] * 1,
    c=lang_counts_top10['Frequency'],
    cmap='plasma',
    alpha=0.7,
    edgecolors='black'
)

plt.title('Top 10 Technology Preferences by Age Group\n(Bubble Size = Frequency)', fontsize=16)
plt.xlabel('Programming Language', fontsize=12)
plt.ylabel('Age Group', fontsize=12)
plt.xticks(rotation=45, ha='right')
cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents', fontsize=12)
plt.tight_layout()
plt.grid(True)
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
import matplotlib.pyplot as plt

# Select relevant columns and drop missing values
df_data = df[['DatabaseWantToWorkWith', 'JobSat', 'Age']].dropna().copy()

# Map age groups to a more readable format
age_map = {
    'Under 18 years old': 'Under 18',
    '18-24 years old': '18-24',
    '25-34 years old': '25-34',
    '35-44 years old': '35-44',
    '45-54 years old': '45-54',
    '55-64 years old': '55-64',
    '65 years or older': '65+'
}
df_data['AgeGroup'] = df_data['Age'].map(age_map)

# Split multiple databases into separate rows
df_data['DatabaseList'] = df_data['DatabaseWantToWorkWith'].str.split(';')
df_data = df_data.explode('DatabaseList')

top_data = df_data['DatabaseList'].value_counts().head(10).index

# Group by database and age group to get frequency
data_count = df_data.groupby(['DatabaseList', 'AgeGroup']).size().reset_index(name='Frequency')
data_count_top10 = data_count[data_count['DatabaseList'].isin(top_data)] 

# Plot
plt.figure(figsize=(14, 8))
scatter = plt.scatter(
    x=data_count_top10['DatabaseList'],
    y=data_count_top10['AgeGroup'],
    s=data_count_top10['Frequency'] * 2,  # Adjust size multiplier for better visuals
    c=data_count_top10['Frequency'],
    cmap='viridis',
    alpha=0.7,
    edgecolors='black'
)

# Formatting
plt.title('Preferred Databases by Age Group\n(Bubble Size = Number of Respondents)', fontsize=16)
plt.xlabel('Database', fontsize=12)
plt.ylabel('Age Group', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents', fontsize=12)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Filter and clean data
df_plot = df[['DevType', 'ConvertedCompYearly', 'JobSatPoints_6']].dropna().copy()

df_plot['DevType'] = df_plot['DevType'].str.split(';')
df_plot = df_plot.explode('DevType')
df_plot['DevType'] = df_plot['DevType'].str.strip()

# Step 2: Group data
grouped = df_plot.groupby('DevType').agg({
    'ConvertedCompYearly': 'median',
    'JobSatPoints_6': 'mean',
    'DevType': 'count'  # will be renamed
}).rename(columns={'DevType': 'Respondents'}).reset_index()

# ❌ Fix the typo here: 'Respodents' → ✅ 'Respondents'
top_grouped = grouped.sort_values('Respondents', ascending=False).head(10)

# Step 3: Plotting
plt.figure(figsize=(12, 6))  # Corrected from plt.fig() to plt.figure()
scatter = plt.scatter(
    x=top_grouped['DevType'],
    y=top_grouped['ConvertedCompYearly'],
    s=top_grouped['JobSatPoints_6'] * 200,  # scale size for visibility
    c=top_grouped['JobSatPoints_6'],
    cmap='viridis',
    alpha=0.7,
    edgecolors='w',
    linewidth=1.5
)

plt.xticks(rotation=45, ha='right')
plt.ylabel('Median Yearly Compensation (USD)')
plt.title('Bubble Plot: Compensation Across Developer Roles\n(Bubble Size = Job Satisfaction)')
plt.colorbar(scatter, label='Average Job Satisfaction')

plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here

df_plot = df[['NEWCollabToolsHaveWorkedWith', 'Age']].dropna().copy() 
age_map = {
    'Under 18 years old': 'Under 18',
    '18-24 years old': '18-24',
    '25-34 years old': '25-34',
    '35-44 years old': '35-44',
    '45-54 years old': '45-54',
    '55-64 years old': '55-64',
    '65 years or older': '65+'
}
df_plot['AgeGroup'] = df_plot['Age'].map(age_map) 

df_plot['NEWCollabToolsHaveWorkedWith'] = df_plot['NEWCollabToolsHaveWorkedWith'].str.split(';')  
df_plot = df_plot.explode('NEWCollabToolsHaveWorkedWith')  
df_plot['NEWCollabToolsHaveWorkedWith'] = df_plot['NEWCollabToolsHaveWorkedWith'].str.strip() 

grouped = df_plot.groupby(['NEWCollabToolsHaveWorkedWith', 'AgeGroup']).size().reset_index(name ='Frequency')

top_grouped = grouped.sort_values('Frequency', ascending = False).head(10)  

plt.figure(figsize = (12,6)) 
scatter = plt.scatter(
    x=top_grouped['NEWCollabToolsHaveWorkedWith'] + " (" + top_grouped['AgeGroup'] + ")",
    y=top_grouped['Frequency'],
    s=top_grouped['Frequency'] * 10,  # Scale bubble size
    c=top_grouped['Frequency'],
    cmap='coolwarm',
    alpha=0.7,
    edgecolors='w',
    linewidth=1.2
)

plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency of Usage')
plt.title('Bubble Plot: Collaboration Tools Used by Age Group\n(Bubble Size = Frequency)')
plt.colorbar(scatter, label='Usage Frequency')

plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
df_plot = df[['WebframeWantToWorkWith', 'JobSatPoints_6']].dropna().copy() 

df_plot['WebframeWantToWorkWith'] = df_plot['WebframeWantToWorkWith'].str.split(';') 
df_plot = df_plot.explode('WebframeWantToWorkWith') 
df_plot['WebframeWantToWorkWith']  = df_plot['WebframeWantToWorkWith'].str.strip() 

grouped = df_plot.groupby('WebframeWantToWorkWith').agg({ 
    'JobSatPoints_6' : 'mean', 
    'WebframeWantToWorkWith' : 'count' 
}).rename(columns ={"WebframeWantToWorkWith": "Respondents"}).reset_index() 
top_grouped = grouped.sort_values('Respondents', ascending=False).head(10)

# Step 4: Create the bubble plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(
    x=top_grouped['WebframeWantToWorkWith'],
    y=top_grouped['JobSatPoints_6'],
    s=top_grouped['Respondents'] * 10,  # Scale bubble size
    c=top_grouped['JobSatPoints_6'],    # Color by satisfaction
    cmap='coolwarm',
    alpha=0.7,
    edgecolors='w',
    linewidth=1.5
)

# Step 5: Add labels
plt.xticks(rotation=45, ha='right')
plt.ylabel('Average Job Satisfaction')
plt.title('Preferred Web Frameworks vs. Job Satisfaction\n(Bubble Size = Number of Respondents)')
plt.colorbar(scatter, label='Job Satisfaction')

plt.tight_layout()
plt.show()
```
