### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# !pip install seaborn 
# import seaborn as sns
df_clean = df[['ConvertedCompYearly', 'Age']].dropna()
age_order = [
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
df_clean = df_clean[df_clean['Age'].isin(age_order)]

median = df_clean.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Sort the age groups
median['Age'] = pd.Categorical(median['Age'], categories=age_order, ordered=True)
median = median.sort_values('Age')

plt.figure(figsize = (12,6)) 
plt.plot(median['Age'], median['ConvertedCompYearly'], marker='o')
plt.title("Median Yearly Compensation by Age Group", fontsize =14) 
plt.xlabel("Age Group") 
plt.ylabel("Median Compensation")  
plt.xticks(rotation =45) 
plt.grid(True) 
plt.tight_layout() 
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
df_clean = df[['ConvertedCompYearly', 'Age']] 
age_order =[
    'Under 18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '65 years or older'
]
target_ages = ['25-34 years old', '35-44 years old'] 
df_filtered = df_clean[df_clean['Age'].isin(target_ages)] 
median = df_filtered.groupby('Age')['ConvertedCompYearly'].median().reset_index() 

median['Age'] = pd.Categorical(median['Age'], categories =age_order, ordered = True)
median.sort_values('Age') 

plt.figure(figsize = (10,6)) 
plt.plot(median['Age'], median['ConvertedCompYearly'], marker ='o')
plt.title("Median Yearly Compensation (Ages 25–45)", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Median Compensation (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
df_clean = df[['YearsCodePro', 'JobSatPoints_6']].dropna()

# Replace special string values with numeric
df_clean['YearsCodePro'] = df_clean['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})

# Convert to float (will raise error if any other non-numeric strings exist)
df_clean['YearsCodePro'] = df_clean['YearsCodePro'].astype(float)

# Group by experience and get average job satisfaction
df_grouped = df_clean.groupby('YearsCodePro')['JobSatPoints_6'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df_grouped['YearsCodePro'], df_grouped['JobSatPoints_6'], marker='o')
plt.title("Job Satisfaction vs. Years of Professional Coding Experience", fontsize=14)
plt.xlabel("Years of Professional Experience")
plt.ylabel("Average Job Satisfaction (JobSatPoints_6)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
df_clean = df[['WorkExp', 'ConvertedCompYearly']].dropna() 
df_clean['WorkExp'] = df_clean['WorkExp'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51,
}) 
df_clean['WorkExp'] = df_clean['WorkExp'].astype(float) 

df_grouped = df_clean.groupby('WorkExp')['ConvertedCompYearly'].median().reset_index() 

plt.figure(figsize=(12, 6))
plt.plot(df_grouped['WorkExp'], df_grouped['ConvertedCompYearly'], marker='o', color='teal')
plt.title("Median Yearly Compensation by Years of Experience", fontsize=14)
plt.xlabel("Years of Experience")
plt.ylabel("Median Compensation (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
df_clean =df[['JobSatPoints_6', 'WorkExp']].dropna() 

df_clean['WorkExp'] = df_clean['WorkExp'].replace({ 
    'Less than 1 year': 0.5,
    'More than 50 years': 51
}) 
df_clean['WorkExp'] = df_clean['WorkExp'].astype(float) 

df_grouped = df_clean.groupby('WorkExp')['JobSatPoints_6'].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(df_grouped['WorkExp'], df_grouped['JobSatPoints_6'], marker='o', color='green')
plt.title("Job Satisfaction Across Experience Levels", fontsize=14)
plt.xlabel("Years of Experience")
plt.ylabel("Average Job Satisfaction (JobSatPoints_6)")
plt.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```
