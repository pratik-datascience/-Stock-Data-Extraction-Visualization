### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
df_clean = df[['DatabaseWantToWorkWith']].dropna().copy() 
df_clean['DatabaseWantToWorkWith'] = df_clean['DatabaseWantToWorkWith'].str.split(';')
df_exploded = df_clean.explode('DatabaseWantToWorkWith')

df_exploded['DatabaseWantToWorkWith'] = df_exploded['DatabaseWantToWorkWith'].str.strip() 

database_counts = df_exploded['DatabaseWantToWorkWith'].value_counts().head(5)

plt.figure(figsize = (12,6))
plt.pie(database_counts ,labels = database_counts.index, autopct = '%1.1f%%', startangle = 180)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()


```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
# print(df['DevType'].unique())
df_clean = df[['DevType']].dropna().copy() 
df_clean['DevType'] = df_clean['DevType'].str.split(';') 

df_exploded = df_clean.explode('DevType')  
df_exploded['DevType'] = df_exploded['DevType'].str.strip() 

developer_counts = df_exploded['DevType'].value_counts().head(5) 

plt.figure(figsize = (12,6))
plt.pie(developer_counts ,labels = developer_counts.index, autopct = '%1.1f%%', startangle = 180)
plt.title('Top 5 developer Respondents Want to Work With',loc ='right')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()

```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here


# Step 1: Clean and prepare data
df_clean = df[['LanguageAdmired']].dropna().copy()

# Step 2: Split semicolon-separated admired languages
df_clean['LanguageAdmired'] = df_clean['LanguageAdmired'].str.split(';')

# Step 3: Explode lists into separate rows
df_exploded = df_clean.explode('LanguageAdmired')

# Step 4: Strip whitespace from each language name
df_exploded['LanguageAdmired'] = df_exploded['LanguageAdmired'].str.strip()

# Step 5: Count top 5 admired languages
admired_counts = df_exploded['LanguageAdmired'].value_counts().head(5)

# Step 6: Plot pie chart
plt.figure(figsize=(10, 6))
plt.pie(admired_counts, labels=admired_counts.index, autopct='%1.1f%%', startangle=180)
plt.title('Top 5 Most Admired Programming Languages Among Respondents', loc='center')
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle
plt.show()

```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here

# Find likely matches


# Step 1: Select and clean data
df_clean = df[['AIToolCurrently Using']].dropna().copy()

# Step 2: Split multi-tool entries
df_clean['AIToolCurrently Using'] = df_clean['AIToolCurrently Using'].str.split(';')

# Step 3: Explode list into individual rows
df_exploded = df_clean.explode('AIToolCurrently Using')
df_exploded['AIToolCurrently Using'] = df_exploded['AIToolCurrently Using'].str.strip()

# Step 4: Count top 5 most used AI tools
ai_tool_counts = df_exploded['AIToolCurrently Using'].value_counts().head(5)

# Step 5: Plot pie chart
plt.figure(figsize=(10, 6))
plt.pie(ai_tool_counts, labels=ai_tool_counts.index, autopct='%1.1f%%', startangle=180)
plt.title('Top 5 AI Tools Currently Used by Developers', loc='center')
plt.axis('equal')  # Ensures the pie is circular
plt.show()

```
