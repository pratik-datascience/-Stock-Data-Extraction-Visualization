### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
duplicates =df[df.duplicated(keep = False)] 
print("Duplicate Rows:") 
print(duplicates)
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
df = df.drop_duplicates()
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum() 
print("Missing values per columns:") 
print(missing_values[missing_values> 0]) 
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
remote_missing = df['RemoteWork'].isnull().sum()
print(f"Missing_rows in RemoteWork column: {remote_missing}") 
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
print("Value counts for RemoteWork") 
print(df['RemoteWork'].value_counts(dropna =False)) 
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code 
most_common = df['RemoteWork'].mode()[0] 
print(f"Most frequent value in RemoteWork : {most_common}")
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
df['RemoteWork'].fillna(most_common, inplace = True) 
print("Missing value after imputation:", df['RemoteWork'].isnull().sum())
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
## Write your code here
# Look for columns with 'Compensation' or 'Salary' in their name
comp_columns = [col for col in df.columns if 'Compensation' in col or 'Salary' in col]

# Describe statistics for these columns
for col in comp_columns:
    print(f"\nSummary statistics for '{col}':")
    print(df[col].describe())
```
