##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
df_clean = df['ConvertedCompYearly'].dropna() 
df_clean = df_clean[df_clean < df_clean.quantile(0.95)] # remove top 5% for clear plot 

sns.set(style ='whitegrid')
plt.figure(figsize = (12,6)) 
sns.histplot(df_clean, kde = True, bins = 40, color = 'skyblue') 
plt.title(' Distribution of ConvertedCompYearly ', fontsize = 18) 
plt.xlabel('Yearly Compensation', fontsize = 12)
plt.ylabel('Number of respondents', fontsize = 12) 
plt.tight_layout() 
plt.show() 
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Check and clean string values just in case
df['Employment'] = df['Employment'].str.strip()

# Filter full-time employees
full_time = df[df['Employment'] == 'Employed, full-time']

# Check how many non-null salary values exist
print("Valid salary values:", full_time['ConvertedCompYearly'].notna().sum())

# Calculate mean safely
median_salary = full_time['ConvertedCompYearly'].median()
print(f"Median Compensation: ${median_salary:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
df =df[['Country', 'ConvertedCompYearly', 'Employment']] 
df= df[(df['Employment'] =='Employed, full-time') & (df['ConvertedCompYearly'].notna())]

top_countries = df['Country'].value_counts().head().index 
df =df[df['Country'].isin(top_countries)] 

df =df[df['ConvertedCompYearly'] < df['ConvertedCompYearly'].quantile(0.95)] 
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Country', y='ConvertedCompYearly', palette='Set2')
plt.xticks(rotation=45)
plt.title('Yearly Compensation by Country (Full-Time Employees)', fontsize=16)
plt.ylabel('Yearly Compensation (USD)', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here


df =df[df['ConvertedCompYearly'].notna()] 
lower_bound = df['ConvertedCompYearly'].quantile(0.05) 
upper_bound = df['ConvertedCompYearly'].quantile(0.95)
df_refined = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)] 
df_refined = df_refined.reset_index(drop = True) 
print(df_refined[['ConvertedCompYearly']].describe())   
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")
df = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']]
df = df.dropna() 
correlation_matrix =df.corr()  
print(correlation_matrix) 

plt.figure(figsize = (12,6))
sns.heatmap(correlation_matrix, annot= True , cmap ='coolwarm', fmt=".2f", linewidths = 0.5) 
plt.title("Correlation Heatmap: Compensation, Work Experience, Job Satisfaction")
plt.show()

```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
sns.set(style = 'whitegrid')
plt.figure(figsize = (12, 6) ) 
plt.subplot(1, 2, 1)
sns.scatterplot(data= df, x = 'WorkExp', y = 'ConvertedCompYearly', color= 'blue', alpha =0.6)
plt.title('Compensation vs Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('Yearly Compensation (USD)')

plt.subplot(1,2,2)
sns.scatterplot(data =df, x='JobSatPoints_1', y='ConvertedCompYearly', color ='green', alpha =0.6)
plt.title('Compensation vs Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Yearly Compensation (USD)')

plt.tight_layout()
plt.show()
```
