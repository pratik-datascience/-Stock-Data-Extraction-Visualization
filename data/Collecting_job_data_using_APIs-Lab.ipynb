You can also view  the json file contents  from the following <a href = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json">json</a> URL.



```python
import requests

api_url = "http://127.0.0.1:5000/data"

def get_number_of_jobs_T(technology):
    # Define query parameters
    params = {'Key Skills': technology}

    # Send GET request to the API with the parameters
    response = requests.get(api_url, params=params)

    # Convert response to JSON (list of job postings)
    data = response.json()

    # Count the number of jobs found
    number_of_jobs = len(data)

    return technology, number_of_jobs

# Example usage:
tech, count = get_number_of_jobs_T("Python")
print(f"{tech} jobs found: {count}")
```

Calling the function for Python and checking if it works.


```python
# Call the function
technology, count = get_number_of_jobs_T("Python")

# Print the result
print(f"{technology} jobs found: {count}")
```

```python
import requests

api_url = "http://127.0.0.1:5000/data"

def get_number_of_jobs_L(location):
    # Prepare query with location
    params = {'Location': location}
    
    # Send GET request to the Flask API
    response = requests.get(api_url, params=params)
    
    # Parse JSON response
    data = response.json()
    
    # Count how many job postings match the location
    number_of_jobs = len(data)
    
    return location, number_of_jobs

# Example usage: check jobs in 'New York'
location, count = get_number_of_jobs_L("New York")
print(f"{location} jobs found: {count}")


```

Find the number of jobs postings for each of the technology in the above list.
Write the technology name and the number of jobs postings into the excel spreadsheet.


```python


api_url = "http://127.0.0.1:5000/data"

# Step 3: Function to Get Job Count by Location
def get_number_of_jobs_L(location):
    params = {'Location': location}
    response = requests.get(api_url, params=params)
    if response.status_code == 200:
        return len(response.json())
    else:
        return 0

# Step 4: Collect results
results = []

for loc in locations:
    count = get_number_of_jobs_L(loc)
    results.append({'Location': loc, 'Job Postings': count})

# Step 5: Write to Excel
df = pd.DataFrame(results)
df.to_excel('job_postings_by_location.xlsx', index=False)

print("Excel file created: job_postings_by_location.xlsx")

```

#### In the similar way, you can try for below given technologies and results  can be stored in an excel sheet.


Collect the number of job postings for the following languages using the API:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB


```python
# your code goes here
import requests
from openpyxl import Workbook

technologies = [
    "C", "C#", "C++", "Java", "JavaScript", "Python",
    "Scala", "Oracle", "SQL Server", "MySQL Server",
    "PostgreSQL", "MongoDB"
]

api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/" \
          "IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"


response = requests.get(api_url)
data = response.json()
def get_numbers_of_jobs_T(technology):
    count = 0
    for job in data:
        title = job.get('title', '').lower()
        description = job.get('description', '').lower()
        if technology.lower() in title or technology.lower() in description:
            count += 1
    return technology, count

# Create Excel workbook
wb = Workbook()
ws = wb.active
ws.title = "Language Job Posting"
ws.append(["Technology", "Number of Job Postings"])

# Append job data
for tech in technologies:
    tech_name, job_count = get_numbers_of_jobs_T(tech)
    ws.append([tech_name, job_count])

# Save file
wb.save("Language Job Posting.xlsx")

```
